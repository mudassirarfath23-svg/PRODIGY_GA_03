```python
import markovify

text = """
Artificial Intelligence is transforming industries by automating tasks and improving decision making.
Machine Learning enables computers to learn from data without explicit programming.
Python is widely used for Artificial Intelligence and Data Science projects.
Cloud Computing provides scalable and reliable computing resources.
DevOps improves collaboration between development and operations teams.
Generative AI creates text, images, music, and code from user prompts.
Technology is evolving rapidly and creating new career opportunities.
Learning AI and Cloud Computing helps students build successful careers.
"""

text_model = markovify.Text(text)

for i in range(5):
    sentence = text_model.make_sentence(tries=100)
    print(sentence)
```

```python
for i in range(5):
    sentence = text_model.make_short_sentence(max_chars=100)

    if sentence:
        print(f"{i+1}. {sentence}")
```
